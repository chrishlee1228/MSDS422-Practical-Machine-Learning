```python
import numpy as np
import pandas as pd
import seaborn as sns
import pandas_profiling
train = pd.read_csv('train.csv')
train['Sex2'] = np.where(train['Sex'] == "male", 0, 1)
pandas_profiling.ProfileReport(train)
```

```python
train.info()
train.describe()
```

```python
sns.countplot(data = train, x = 'Sex', hue = 'Survived')
pd.crosstab(train['Sex'],train['Survived'])
```

```python
train['age.group'] = pd.cut(train['Age'],bins = [0,12,17,35,50,100],labels=['0-12','13-17','18-35','36-50', '51-100'])

sns.countplot(data = train, x = 'age.group', hue = 'Survived')
pd.crosstab(train['age.group'],train['Survived'])
```
